```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
print(os.getcwd())
```

```python
final_df=pd.DataFrame()
for file_name in os.listdir('Sales_Data'):
    path= os.path.join('Sales_Data',file_name)
    df = pd.read_csv(path)
    final_df = pd.concat([final_df,df])
print(final_df.shape)
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df=final_df[final_df['Quantity Ordered']!='Quantity Ordered']
```

```python
# final_df['month']=final_df['Order Date'].str[0:2]
final_df['Quantity Ordered']=pd.to_numeric(final_df['Quantity Ordered'])
final_df['Price Each']=pd.to_numeric(final_df['Price Each'])
final_df['sales']=final_df['Quantity Ordered']*final_df['Price Each']
final_df['date']=pd.to_datetime(final_df['Order Date'])
final_df['month']=final_df['date'].dt.month
final_df.head()
```

```python
final_df['city']=final_df['Purchase Address'].apply(lambda x: f"{x.split(',')[1]}({x.split(',')[2].split()[0]})")
final_df.head(20)
```

```python
final_df['city'].unique()
```

```python
final_df['hour']=final_df['date'].dt.hour
final_df.head()
```

```python
final_df.head(20)
```

```python
print(final_df.shape)
dup_df = final_df[final_df['Order ID'].duplicated(keep=False)]
dup_df.head()
```

```python
final_df.info()
```

```python
final_df.groupby('month').sum()['sales']
```

```python
month=[key for key,df in final_df.groupby('month')]
plt.bar(month,final_df.groupby('month').sum()['sales'])
plt.xlabel('Month')
plt.ylabel('sales')
plt.yticks(final_df.groupby('month').sum()['sales'])
plt.xticks(month)
plt.show()
```

```python
city=[key for key,df in final_df.groupby('city')]
plt.bar(city,final_df.groupby('city').sum()['sales'])
plt.xlabel('city')
plt.ylabel('sales')
plt.yticks(final_df.groupby('city').sum()['sales'])
plt.xticks(city,rotation=90,size=8)
plt.show()
```

```python
plt.figure(figsize=(12,7))
hour=[key for key,df in final_df.groupby('hour')]
plt.plot(hour,final_df.groupby('hour').sum()['sales'])
plt.xlabel('hour')
plt.ylabel('sales')
#plt.yticks(final_df.groupby('hour').sum()['sales'])
plt.xticks(hour,rotation=0,size=8)
plt.show()
```

```python
for key,df in dup_df.groupby('Order ID'):
    print(key)
    products = ",".join(df['Product'])
    dup_df.loc[dup_df['Order ID']==key,'Product']=products
    #dup_df.loc[]
```

```python
copy_dup_df=dup_df.drop_duplicates(subset=['Order ID'],keep='first')
```

```python
print(dup_df.shape)
print(copy_dup_df.shape)
```

```python
from itertools import combinations
from collections import Counter

counter=Counter()
for index,row in copy_dup_df.iterrows():
    product_list = row['Product'].split(',')
    c=combinations(product_list, 2)
    counter.update(c)
    #print(count)
```

```python
print(counter.most_common(100))
```

```python
relations = pd.DataFrame(counter.most_common(100))
relations[0]=relations[0].apply(lambda x: ",".join(x))
```

```python
relations.head()
```

```python
plt.figure(figsize=(30,15))
plt.bar(relations[0],relations[1])
plt.xlabel('products')
plt.ylabel('no of sales')
#plt.yticks(final_df.groupby('city').sum()['sales'])
plt.xticks(relations[0],rotation=90,size=8)
plt.show()
```
